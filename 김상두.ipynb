```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

line = int(len(df) * 0.8)
df = df.iloc[:line]

std1 = df['f1'].std()

med=df['f1'].median()
df['f1'] = df['f1'].fillna(med)

std2 = df['f1'].std()

result = abs(std1-std2)
print(round(result,2))
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

std = df['age'].std()
mean = df['age'].mean()

lower = mean - (std * 1.5)
upper = mean + (std * 1.5)

cond1 = df['age'] < lower
cond2 = df['age'] > upper

print(df[cond1|cond2]['age'].sum())
```
